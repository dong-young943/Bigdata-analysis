#### 데이터정렬과 평균치 산출
- alcohol(음주수치)이 상위 10번째 값으로 상위 1~10위의 범위의 값을 변경한 후
- speeding(속도)가 7이상의 alcohol(음주수치) 데이터들의 평균 산출
- 최종 결과만 print 출력:
###### seaborn, pandas 라이브러리불러오기

```python
import seaborn as sns
import pandas as pd
sns.get_dataset_names()
```

###### seaborn의 'car_crashes'샘플 데이터 활용

```python
df=sns.load_dataset('car_crashes')
print(df.head())
```

###### alcohol열을 내림차순으로 정렬한 후 data 변수에 할당

```python
data=df.sort_values(by=df.columns[2], ascending=False)
data.head(10)
```

###### alcohol열의 상위 10번째 값을 추출하여 ten 변수에 할당

```python
ten = data.iloc[9,2]
print(ten)
```

###### alcohol열의 상위 1~10위 범위의 값을 추출하여 ten 변수의 값으로 대체

```python
data.iloc[:10, 2] = ten
data.head(10)
```

###### 수정된 데이터를 기반으로 speeding열이 7이상인 값을 result 변수에 할당
###### mean함수로 alcohol열의 평균값 산출, print 출력

```python
result=data[(data['speeding']>=7)]
print('최종 결과물은',result['alcohol'].mean())
```
